```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.impute import SimpleImputer
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('J:/Atieh/University/data mining/homeworks/HW2/code/q6/train.csv')
label = df.Survived
df = df.drop('Survived', axis=1)
df = df.drop('PassengerId', axis = 1)
df = df.drop('Name', axis=1)
df = df.drop('Ticket', axis=1)
df = df.drop('Cabin', axis=1)
# decision tree works only with numbers so
# change male to 0, female to 1
df = df.replace('male', 0)
df = df.replace('female', 1)
# change destinations' type to numbers
df.Embarked = df.Embarked.replace('C', 1)
df.Embarked = df.Embarked.replace('Q', 2)
df.Embarked = df.Embarked.replace('S', 3)
# features_train = features_train.replace('S', 2)
# features_train = features_train[].replace('Q', 3)


features_train, features_test, label_train, label_test = train_test_split(df, label, test_size = 0.2)

```

```python
features_train.Age = features_train.Age.fillna(features_train.Age.mean())
mean_age = features_train.Age.mean()
mean_fare = features_train.Fare.mean()

s = features_train.Embarked.mode()


features_train.Embarked = features_train.Embarked.fillna(float(s))


features_train
```

```python
features_train.isnull().sum()
```

```python
features_train
```

```python
DecisionTree_Class_Model = DecisionTreeClassifier( criterion= 'gini', max_depth=6)
DecisionTree_Class_Model = DecisionTree_Class_Model.fit(features_train, label_train)
DecisionTree_Class_Model
```

```python
features_test
```

```python
features_test.Age = features_test.Age.fillna(mean_age)
features_test.Fare = features_test.Fare.fillna(mean_fare)
features_test.Embarked = features_test.Embarked.fillna(float(s))
features_test.isnull().sum()
```

```python
label_pred = DecisionTree_Class_Model.predict(features_test)
label_pred
```

```python
accuracy = metrics.accuracy_score(label_test, label_pred)
accuracy
```
